## Video processing pipeline

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture('vids/test.mp4')

if not cap.isOpened():
    print("Error: Could not open video.")
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        # OpenCV uses BGR, Matplotlib uses RGB. We must convert!
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        plt.imshow(frame_rgb)
        plt.title("Video Frame")
        plt.axis('off') # Hide X and Y axes
        plt.show()
    else:
        print("Could not read frame.")

cap.release()
```
